Terminologies:<br><br>
a) Distribution: A distribution in statistics is a function that shows the possible values for a variable and how often they occur.<br><br>
b)Categorical Variables: A categorical variable (also called qualitative variable) refers to a characteristic that can't be quantifiable. Categorical variables can be either nominal or ordinal.<br><br>
c) Continuous Variables: A variable is said to be continuous if it can assume an infinite number of real values within a given interval.

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as pyp
%matplotlib inline
import seaborn as sns
```

```python
population = pd.read_csv("D:\PRODIGY\PRODIGY_DS_01\world_population.csv")
population.head()
```

<b>Using Seaborn Library</b>

```python
sns.histplot(data = population, x = "Continent", kde = True)
```

<b>Using Matplotlib Library</b>


```python
df = pd.DataFrame(population)
Continent = df['Continent']
pop = df['2022 Population']
pyp.bar(Continent, pop, width=0.25, color="red")
pyp.show()
```

```python
pyp.hist(Continent, bins=15, rwidth=1000 , color="purple")
```
